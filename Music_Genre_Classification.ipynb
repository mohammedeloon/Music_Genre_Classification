```python
from sklearn.feature_selection import VarianceThreshold, SelectKBest

var_th = VarianceThreshold(threshold=0.8).set_output(transform='pandas')
selected_features = var_th.fit_transform(m_data_nums)
```

```python
selected_features.head()
```

## splitting data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(selected_features, music_labels, test_size=0.25,
                                                random_state=42, shuffle=True, stratify=music_labels)
```
